```python
import pandas as pd
```

```python
data=pd.read_csv('Netflix_Dataset.csv')
```

```python
# 1. Load the Netflix dataset and display the first five records.
data.head()
```

```python
# 2. How many unique movies and TV shows are in the dataset?
data['type'].value_counts()
```

```python
# 3. What is the total number of missing values in each column?
data.isnull().sum()
```

```python
# 4. Display the top 5 rows of the dataset.
data.head()
```

```python
# 5. What are the different genres available in the dataset? (unique genres)
data['genre'].unique()
```

```python
# 5. What are the different genres available in the dataset? (count of unique genres)
data['genre'].nunique()
```

```python
# 6. How many unique directors are there in the dataset?
data['director'].nunique()
```

```python
# 6. Find all movies released in 2019.
data[ (data['type']=='Movie') & (data['release_year']==2019) ]
```

```python
# 7. Retrieve all TV Shows directed by Neeraj Pandey.
data[(data['type']=='Movie') & (data['director']=='Neeraj Pandey') ]
```

```python
# 8. List all the movies where Akshay Kumar is part of the cast.
data[(data['type'] == 'Movie') & (data['cast'].str.contains('Akshay Kumar'))]
```

```python
# 9. Show all content that has a rating of UA and a duration of more than 150 minutes.

type_movie=pd.DataFrame(data[data['type']=='Movie'])
type_movie[type_movie['duration'].str.replace('min','').astype(int)>150]
```

```python
# 10. Find all content where the genre is either Comedy or Drama.
data[(data['genre']=='Comedy') | (data['genre']=='Drama') ]
```

```python
# 11. Sort the dataset by release year in descending order.
data.sort_values(['release_year'], ascending=False)
```

```python
# 12. Sort the dataset first by type and then by title alphabetically.
data.sort_values(['type','title'],ascending=[True,True])
```

```python
# 13. Find the number of movies and TV shows released per year.
data.groupby(['release_year','type']).size()
```

```python
# 14. What are the top 3 most common genres and their counts?
data['genre'].value_counts().head(3)
```

```python
# 15. Find the number of movies directed by each director.
data[data['type']=='Movie'].groupby(['director']).size().sort_values()
```
