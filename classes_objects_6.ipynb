```python
# Write a python program to calculate prime numbers between n1 and n2 using classes and objects
class isPrime:
    def __init__(self, n1, n2):
        self.n1 = n1
        self.n2 = n2
    def pPrime(self):
      p=0
      for i in range (self.n1,self.n2+1):
        for j in range(2,i):
          if i%j==0:
            p=1
            break
        if p==0:
          print(i)
        else:
          p=0
pme = isPrime(1,20)  #This will call the __init__ function
pme.pPrime()          #pme is called and stored in self <- implicitly done
```

```python
class Prime:
  def __init__(self, n1, n2):
    self.n1 = n1
    self.n2 = n2

  def isPrime(self):
    p = 0
    for i in range(self.n1,self.n2+1):
      for j in range(2,i):
        if i%j==0:
          p = 1
          break

      if p==0:
        print(i)
      else:
        p=0

p1= Prime(1,15)
p1.isPrime()
```

```python
class Bank:
    def __init__(self, accid, accname, bal):
        self.accid = accid
        self.accname = accname
        self.bal = bal

    def deposit(self, amount):
        self.bal += amount
        print(f"Rs {amount} deposited successfully.")

    def withdraw(self, amount):
        if self.bal >= amount:
            self.bal -= amount
            print(f"Rs {amount} withdrawn successfully.")
        else:
            print("Insufficient balance")

    def display(self):
        print("Current Balance = Rs", self.bal)


accounts = {}

while True:
    print("\nMenu:")
    print("1. Create Account")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Display Balance")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        accid = input("Enter account ID:\t")
        accname = input("Enter account name:\t")
        bal = int(input("Enter initial balance:\t"))
        accounts[accid] = Bank(accid, accname, bal)
        print("Account created successfully.")

    elif choice in ('2', '3', '4'):
        accid = input("Enter account ID:\t")
        if accid in accounts:
            account = accounts[accid]

            if choice == '2':
                amount = int(input("Enter amount to deposit: Rs "))
                account.deposit(amount)

            elif choice == '3':
                amount = int(input("Enter amount to withdraw: Rs "))
                account.withdraw(amount)

            elif choice == '4':
                account.display()
        else:
            print("Account not found.")

    elif choice == '5':
        print("Exiting...")
        break

    else:
        print("Invalid choice. Please try again.")
```
