```python
ecom=pd.read_csv('E-Commerce.csv')
```

```python
# 3. List all the column names in the dataset.
ecom.columns
```

```python
# 4. Check if there are any missing values in the dataset.
ecom.isnull().sum()
```

```python
# 5. Display unique values in the 'Category' column.
ecom['Category'].unique()
```

```python
ecom['Category'].nunique()
```

```python
# 6. Find the total number of products sold.
ecom['Quantity'].sum()
```

```python
# 7. Calculate the average price of products in each category.
ecom.groupby('Category')['Price'].mean()
```

```python
# 8. Display the details of the most expensive product.
ecom[ecom['Price']==ecom['Price'].max()]
```

```python
# 8. Display the details of the most expensive product.
ecom.loc[ecom['Price'].idxmax()]
```

```python
# 9. Find the customer who made the highest total purchase.
ecom['TotalQuantity']=ecom['Price']*ecom['Quantity']
# A new column 'TotalQuantity' is added in the dataset
```

```python
# 9. Find the customer who made the highest total purchase.
ecom.groupby('Customer_ID')['TotalQuantity'].sum().sort_values(ascending=False).head(1)
```

```python
# # 10.Calculate the total revenue generated by the e-commerce store.
ecom['TotalQuantity'].sum()
```

```python
# 11.Show the products ordered on a specific date, e.g., '2024-01-05'.
ecom[ecom['Order_Date']=='05-01-2024']
```

```python
# 12.Create a new column 'Total_Cost' (Price * Quantity) and display the dataset.
ecom['Total_Cost']=ecom['Price']*ecom['Quantity']
ecom
```

```python
# 13.Find the top 3 customers who spent the most.
ecom.groupby('Customer_ID')['Total_Cost'].sum().sort_values(ascending=False).head(3)
```

```python
# 14.Count the number of orders for each category.
ecom.groupby('Category')['Order_ID'].count()
```

```python
# 15.Display the products with a quantity greater than 3.
ecom[ecom['Quantity']>3]
```

```python
# 16.Calculate the average quantity of products sold per order.
ecom.groupby('Order_ID')['Quantity'].mean()
```

```python
# 17.Show the products with a price less than $50.
ecom[ecom['Price']<50]
```

```python
# 18.Find the most popular product (highest total quantity sold).
ecom.groupby('Order_ID')['TotalQuantity'].sum().sort_values(ascending=False).head(1)
```

```python
# 19.Display the orders made by a specific customer, e.g., 'C105'.
ecom[ecom['Customer_ID']=='C105']
```

```python
# 20.Calculate the total cost of products in each category.
ecom.groupby('Category')['Total_Cost'].sum()
```

```python
# 21.Display the products that were ordered more than once.
order=ecom['Product_Name'].value_counts()
order[order>1]
```

```python
# 22.Create a new Data Frame that contains only Electronics products.
df=pd.DataFrame(ecom)
df[df['Category']=='Electronics']
```

```python
# 23.Find the date with the highest number of orders.
ecom['Order_Date'].value_counts().sort_values(ascending=False).head(1)
```
